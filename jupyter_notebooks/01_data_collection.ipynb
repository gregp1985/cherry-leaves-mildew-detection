## Dataset Description

The dataset consists of images of cherry leaves collected by Farmy & Foods.

There are two classes:
- Healthy: images of cherry leaves without disease
- Powdery Mildew: images of cherry leaves affected by fungal infection

Each image is a colour image with a resolution of 256x256 pixels.

This dataset will be used for:
- Visual analysis to differentiate healthy vs infected leaves
- Training a machine learning model to classify leaf health status
